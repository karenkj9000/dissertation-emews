```python
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
df = pd.read_excel("EMEWS.xlsx", header=3)
```

```python
df.shape
```

```python
df.columns
```

## Box Plots

```python
max_cols = 3
num_plots = len(df.columns)
ncols = min(max_cols, num_plots)
nrows = math.ceil(num_plots / ncols)

# Create box plots for all columns
df.plot(kind='box', subplots=True, layout=(nrows, ncols), figsize=(5 * ncols, 4 * nrows), sharex=False, sharey=False)
plt.tight_layout()
plt.show()
```

```python
lower_col = df['Day '].str.lower()

# Identify rows that start with "total emews" (in lowercase now)
mask = lower_col.str.startswith("total emews", na=False)
```

```python
df[mask]
```

```python
df_without_total = df[~mask]
```

```python
max_cols = 3
num_plots = len(df.columns)
ncols = min(max_cols, num_plots)
nrows = math.ceil(num_plots / ncols)

# Create box plots for all columns
df_without_total.plot(kind='box', subplots=True, layout=(nrows, ncols), figsize=(5 * ncols, 4 * nrows), sharex=False, sharey=False)
plt.tight_layout()
plt.show()
```
